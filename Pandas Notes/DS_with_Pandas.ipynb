# Creating new columns

```python
# df['Total'] = df['Attack'] + df['Defense']
df['Total'] = df.iloc[:, 5:7].sum(axis=1)

df.head(4)
```

# Deleting columns

```python
df = df.drop(columns=['Total'])
df.head(4)
```

# Filtering data

```python
df.loc[(df['Name'] == 'Ivysaur') & (df['Type 1'] == 'Grass')]
```

```python
df.loc[~df['Name'].str.contains('Mega')].head(5)
```

```python
df.drop_duplicates(['Type 1'])
```

## Regex (regular expressions)

```python
import re 

new_df = df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)].head(10) # re.I - Ignore Case
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)].head(10)
# [a-z]* means 0 or more occurences of any letter
# ^ means that PI is a begining of each line (cant stand in the middle)
```

```python
df.loc['content'].apply(lambda text: len(re.findall(r'\bWORD\b', text))) # \b - word boundary
```

# Changing data


```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flame'
df.head(15)
```

```python
df.loc[df['Type 1'] == 'Flame', 'Type 1'] = 'Fire'
```

# Datetime 

```python
df['year'] = df.loc['purchase_date'].dt.year
df['month'] = df.loc['purchase_date'].dt.month

df['timeformat'] = df.loc['purchase_date'].dt.strftime('%Y-%m')
```

# Groupby

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```
